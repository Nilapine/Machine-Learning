```python
q1 = kolom_numerik.quantile(0.25)
q3 = kolom_numerik.quantile(0.75)
iqr = q3 - q1
iqr
```

```python
# handle warning 
import warnings 
warnings.filterwarnings('ignore')

# outlier filter
df_noid_align, iqr_new = kolom_numerik.align(iqr, axis=1, copy=False, join='outer')
outlier_filter = (kolom_numerik < q1 - 1.5 * iqr_new) | (kolom_numerik > q3 + 1.5 * iqr_new)
outlier_filter
```

```python
df[outlier_filter['Rating']] \
    .loc[:, ['Model', 'Rating']] \
    .sort_values(by=['Rating'], ascending=False)
```

```python
# memfilter DataFrame df sehingga hanya baris-baris yang memiliki nilai 'core i7' di kolom 'processor_tier' 
processor = df[df['processor_tier'] == 'core i7']
processor
```

```python
# menyorot satu bar yang dipilih berdasarkan nama brand tertentu
df['brand'].value_counts().plot(kind='bar')
# Menghitung value counts
value_counts = df['brand'].value_counts()

# Mendapatkan nama brand yang paling banyak muncul
highlighted_brand = 'lenovo'

# Membuat list warna default untuk setiap bar
colors = ['skyblue'] * len(value_counts)

# Mengubah warna untuk bar yang merupakan highlighted_brand
highlight_index = value_counts.index.get_loc(highlighted_brand)
colors[highlight_index] = 'orange'

# Plot bar chart dengan warna yang telah ditentukan
value_counts.plot(kind='bar', color=colors)

plt.show()
```

```python
# Data
ability = ['amd', 'intel', 'apple', 'other']
score = [705, 267, 15, 4]

# Plot pie chart
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('equal')

# Tentukan highlight
explode = (0, 0.1, 0, 0)  # highlight hanya pada bagian 'apple'

# Plot pie chart dengan ledakan
ax.pie(score, labels=ability, autopct='%1.2f%%', explode=explode)

plt.show()
```
